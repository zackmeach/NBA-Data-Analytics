```python
import nba_api
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
pd.set_option("display.max_columns", None)
```

```python
from nba_api.stats.static.teams import find_team_by_abbreviation
from nba_api.stats.endpoints.teamdashboardbygeneralsplits import TeamDashboardByGeneralSplits
```

```python
cle = find_team_by_abbreviation('CLE')['id']
```

---
## Base

```python
dash = TeamDashboardByGeneralSplits(team_id=cle,
                                    season='2020-21',
                                    measure_type_detailed_defense='Base',
                                    pace_adjust='Y',
                                    per_mode_detailed='Per100Possessions').get_data_frames()[3]
dash.drop(columns=['GROUP_SET', 'SEASON_MONTH_NAME'], inplace=True)
dash.drop(dash.iloc[:, 27:], inplace=True, axis=1)
dash
```
